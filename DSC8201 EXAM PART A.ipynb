```python
# Replace whitespace with NaN
data_new.replace(r'^\s*$', np.nan, regex=True, inplace=True)

# Count NaN values in each column
missing_values = data_new.isna().sum()

print(missing_values)
```

There is missing data for variables DES_GEOUNI,GEO_UNIT ,Climate_Type ,DES_ClimateType, Landuse_Type and SUB_Basin  

<b>How i plan to handle missing data:</b><br>
a) Categorical Data: For categorical variables like Landuse_Type and SUB_Basin, i will impute missing values using the mode (most frequent category) <br>b)
Textual Descriptions: For descriptive text fields like DES_GEOUNI and DES_ClimateType,i will imput  a placeholder like "Unknown".

- A

```python
# Calculating the mode for 'Landuse_Type' and 'SUB_Basin'
landuse_Type_mode = LSFI_data['Landuse_Type'].mode()[0]
sub_basin_mode = LSFI_data['SUB_Basin'].mode()[0]

# Imputing missing values using the mode
LSFI_data['Landuse_Type'].fillna(landuse_Type_mode, inplace=True)
LSFI_data['SUB_Basin'].fillna(sub_basin_mode, inplace=True)

# Check if the imputation worked correctly by counting nulls again in these columns
imputed_missing_values = LSFI_data[['Landuse_Type', 'SUB_Basin']].isna().sum()

landuse_Type_mode, sub_basin_mode, imputed_missing_values
```

- B

```python
# Replace missing values with "Unknown"
LSFI_data['DES_GEOUNI'].fillna('Unknown', inplace=True)
LSFI_data['DES_ClimateType'].fillna('Unknown', inplace=True)

# Verify the imputation
missing_values_after_imputation = LSFI_data[['DES_GEOUNI', 'DES_ClimateType']].isna().sum()
print(missing_values_after_imputation)
```

<b>To generate insights on landslide occurrences in Iran, we can explore:</b>

<font color='Red'>The distribution of landslides in terms of elevation, slope, and proximity to water sources and faults.</font>

```python
# Setting the style of the plots
sns.set(style="whitegrid")

# Creating a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Distribution of Slope (in degrees)
sns.histplot(data=LSFI_data, x='Slop(Degrees)', kde=True, ax=axes[0, 0], bins=30, color='blue')
axes[0, 0].set_title('Distribution of Slope (Degrees)')
axes[0, 0].set_xlabel('Slope (Degrees)')
axes[0, 0].set_ylabel('Frequency')

# Distribution of Elevation
sns.histplot(data=LSFI_data, x='Elevation', kde=True, ax=axes[0, 1], bins=30, color='green')
axes[0, 1].set_title('Distribution of Elevation')
axes[0, 1].set_xlabel('Elevation (m)')
axes[0, 1].set_ylabel('Frequency')

# Distribution of Distance to Nearest River
sns.histplot(data=LSFI_data, x='RiverDIST(m)', kde=True, ax=axes[1, 0], bins=30, color='red')
axes[1, 0].set_title('Distribution of Distance to Nearest River')
axes[1, 0].set_xlabel('Distance to River (m)')
axes[1, 0].set_ylabel('Frequency')

# Distribution of Distance to Nearest Fault
sns.histplot(data=LSFI_data, x='FaultDIST(m)', kde=True, ax=axes[1, 1], bins=30, color='purple')
axes[1, 1].set_title('Distribution of Distance to Nearest Fault')
axes[1, 1].set_xlabel('Distance to Fault (m)')
axes[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

<b>Here are the distributions of several key factors related to landslide occurrences in the dataset:</b>

- Slope (Degrees): The distribution shows a concentration of landslides at lower slope angles, with a significant tail extending to higher angles. This indicates that while landslides occur across a range of slopes, they are more frequent at specific, perhaps moderate to steep, slope angles.
- Elevation: Landslides are distributed across a wide range of elevations, with a concentration at mid-elevations. This suggests that elevation plays a significant role in landslide occurrence.
- Distance to Nearest River: The distribution is right-skewed, meaning many landslides occur close to rivers. 
- Distance to Nearest Fault: Similar to the river proximity, there's a right-skewed distribution showing that a considerable number of landslides occur near geological faults. 

<font color='Red'>Distribution of Landslides Across Different Sub-Basins in Iran</font>

```python
# Count the number of landslides in each sub-basin
sub_basin_landslides = LSFI_data['SUB_Basin'].value_counts()

# Plot the distribution
plt.figure(figsize=(10, 6))
sub_basin_landslides.plot(kind='bar', color='skyblue')
plt.title('Distribution of Landslides Across Different Sub-Basins in Iran')
plt.xlabel('Sub-Basin')
plt.ylabel('Number of Landslides')
#plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The bar chart above illustrates the distribution of landslide occurrences across different sub-basins in Iran. Some sub-basins experience more landslides than others, indicating possible variations in geographical or environmental factors across these areas.and.

<font color='Red'>How different types of land use and climate conditions might influence the frequency of landslides by examining these attributes in the dataset</font>

```python
# Plotting the distribution of landslides by land use type and climate type

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 6))

# Bar plot for land use type
landuse_counts = LSFI_data['Landuse_Type'].value_counts()
landuse_counts.plot(kind='bar', ax=axes[0], color='green')
axes[0].set_title('Distribution of Landslides by Land Use Type')
axes[0].set_xlabel('Land Use Type')
axes[0].set_ylabel('Number of Landslides')
#axes[0].tick_params(axis='x', rotation=45)

# Bar plot for climate type
climate_counts = LSFI_data['Climate_Type'].value_counts()
climate_counts.plot(kind='bar', ax=axes[1], color='orange')
axes[1].set_title('Distribution of Landslides by Climate Type')
axes[1].set_xlabel('Climate Type')
axes[1].set_ylabel('Number of Landslides')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
